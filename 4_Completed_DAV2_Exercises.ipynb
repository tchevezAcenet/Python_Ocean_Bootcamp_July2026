#### Exercise 2: Complete the 'handle_missing_values' function to handle missing values in the dataset. 

Iterate each column of the dataframe **df_filled**. Using **fillna()** fill the ***NaN*** values in each column by replacing it with the average (mean) value if it's a numeric column or with the string "NA" if it's a text column.

*Hint* : You can use the following to detect if the column's data type is a number with the following condition:
-  **if pd.api.types.is_numeric_dtype(df_filled[column])**

***Note:*** Use *df.copy()* when you want to create a completely independent copy of a DataFrame so changes to one do not affect the other.


```python
# Function to handle missing values for both text and numeric columns
def handle_missing_values(df):
    df_filled = df.copy()
    
    for column in df_filled.columns:
        if pd.api.types.is_numeric_dtype(df_filled[column]):
            mean_value = round(df_filled[column].mean(),2)
            df_filled[column] = df_filled[column].fillna(mean_value)
        else:
            df_filled[column] = df_filled[column].fillna('NA')

    return df_filled
```

#### Exercise 3: Complete the 'handle_outliers' function to handle outliers in the dataset
Same as Question 2, iterate each column of the dataframe, detect if the column's data type is a number, if it is then identify outliers using the interquartile range (IQR) method teached in class.  

***Note:*** Use *df.copy()* when you want to create a completely independent copy of a DataFrame so changes to one do not affect the other.

```python
# Function to handle outliers for numeric columns
def handle_outliers(df):
    df_outliers_removed = df.copy()
    
    for column in df_outliers_removed.columns:
         if pd.api.types.is_numeric_dtype(df_outliers_removed[column]):
            q1 = df_outliers_removed[column].quantile(0.25)
            q3 = df_outliers_removed[column].quantile(0.75)
            iqr = q3 - q1
            
            lower_threshold = round(q1 - 1.5 * iqr, 2)
            upper_threshold = round(q3 + 1.5 * iqr, 2)

            number_outliers = df_outliers_removed[
                (df_outliers_removed[column] <  lower_threshold) |
                (df_outliers_removed[column] > upper_threshold)
            ]

            print(f"{column} - Number of existent outliers: {len(number_outliers)}")
            
            df_outliers_removed = df_outliers_removed[((df_outliers_removed[column]>= lower_threshold) 
                                                        & (df_outliers_removed[column] <= upper_threshold))]

            #Check
            remaining_outliers = df_outliers_removed[
                (df_outliers_removed[column] <  lower_threshold) |
                (df_outliers_removed[column] > upper_threshold)
            ]

            print(f"{column} - Check - Number of outliers after: {len(remaining_outliers)}\n")
    
    return df_outliers_removed
```

```python
df_without_outliers = handle_outliers(df)
print("Dataframe Shape:", df_without_outliers.shape)
```

#### Exercise 4: Complete the 'handle_duplicates' function to remove duplicates in the dataset
Drop duplicates from the **df** dataframe and save it in the **df_deduplicated** variable. Print the shape of the original dataset and the shape of the Deduplicated dataset to compare.

```python
# Function to handle duplicates
def handle_duplicates(df):
    # Drop duplicates and save the result in a new DataFrame
    df_deduplicated = df.drop_duplicates()
    
    # Print the number of duplicates removed
    print(f"Original dataset shape: {df.shape}")
    print(f"Deduplicated dataset shape: {df_deduplicated.shape}")
    
    return df_deduplicated
```

#### Exercise 5: Complete the 'standardize_data' function to standardizes the 'Species' column in the dataset
Using the following dataset, standarize the *Species* column.

***Note:*** Use *df.copy()* when you want to create a completely independent copy of a DataFrame so changes to one do not affect the other.

```python
data = {
    'Species': [' blue whale ', 'great White shark', 'dolphin', 'BLUE whale', 'doLPHin'],
    'Count': [5, 2, 13, 4, 11]
}
df_ocean_species = pd.DataFrame(data)
```

```python
# Function to address inconsistency and standardize data
def standardize_data(df):
    df_standardized = df.copy()
    
    # Standardize species names
    df_standardized['Species'] = df_standardized['Species'].str.lower().str.strip().replace(
        {'blue whale': 'Blue Whale',
         'great white shark': 'Great White Shark',
         'dolphin': 'Dolphin'}
    )
    
    return df_standardized
```

```python
# Standardize the species names
df_standardized_species = standardize_data(df_ocean_species)

# Display the standardized DataFrame
print("Original DataFrame:")
print(df_ocean_species)
print("\nStandardized DataFrame:")
print(df_standardized_species)
```
